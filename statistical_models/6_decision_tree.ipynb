```python
import pandas as pd
```

Reading data

```python
df = pd.read_csv('/content/drive/MyDrive/ml_practice/decision_tree/salaries.csv')
df.head(3)
```

Splitting inputs (X) and target (y) from data

```python
X = df.drop('salary_more_then_100k', axis ='columns')
X.head(3)
```

```python
y = df['salary_more_then_100k']
y.head(3)
```

Machine Learning algorithms only work on numbers and not text, so converting all X column/ features to numbers

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
X['company_n'] = le_company.fit_transform(X['company'])
X['job_n'] = le_job.fit_transform(X['job'])
X['degree_n'] = le_degree.fit_transform(X['degree'])
X.head(3)
```

```python
X = X.drop(['company','job','degree'], axis ='columns')
X.head(3)
```

```python
from sklearn import tree
model_dt = tree.DecisionTreeClassifier()
```

```python
model_dt.fit(X, y)
```

```python
model_dt.score(X,y) # shows 1 because it is fed with same data that was trained
```

```python
model_dt.predict([[2,2,1]])
```

```python
df_tit = pd.read_csv('/content/drive/MyDrive/ml_practice/decision_tree/titanic.csv')
df_tit.head(5)
```

Taking only the required inputs to train the model

```python
inputs = df_tit.drop(['PassengerId','Survived','Name','SibSp','Parch','Ticket','Cabin',	'Embarked'], axis='columns')
inputs.head(5)
```

```python
target = df_tit['Survived']
target.head(5)
```

Converting 'Sex' - text column to label encoder

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
inputs['sex_n'] = le.fit_transform(inputs['Sex'])
inputs.head(5)
```

```python
inputs = inputs.drop('Sex', axis='columns')
inputs.head(5)
```

```python
inputs.shape
```

```python
target.shape
```

Splitting the train and test data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size =0.2)
```

```python
from sklearn import tree
model_tit = tree.DecisionTreeClassifier()
```

```python
model_tit.fit(X_train, y_train)
```

```python
y_prediction = model_tit.predict(X_test)
y_prediction
```

```python
model_tit.score(X_test, y_test)
```
