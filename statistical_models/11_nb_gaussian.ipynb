```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/ml_practice/naive_bayes/titanic.csv')
df.head()
```

Taking only required columns needed for dataset

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns', inplace=True)
df.head()
```

Dropping target column for input dataframe

```python
input = df.drop(['Survived'], axis='columns')
target = df['Survived']
df.head()
```

Creating dummy varibales for text columns

```python
dummies = pd.get_dummies(input.Sex, dtype=int)
dummies.head()
```

Merging the dummy variables with original column and dropping the text columns as machine learning algorithms can't perform with text data

```python
input = pd.concat([input,dummies],axis='columns')
input.head()
```

```python
input = input.drop(['Sex'], axis='columns')
input.head()
```

Checking if there is any NaN values

```python
input.columns[input.isna().any()]
```

```python
input['Age'] = input['Age'].fillna(input['Age'].mean())
input.head()
```

```python
target.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(input, target, test_size=0.25)
```

There are 3 types in Navie Bayes:
- Bernoulli Naive Bayes - used when features are binary
- Multinomial Naive Bayes - when features are count or frequencies
- Gaussian Naive Bayes - when features are normally distributed.

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test, y_test)
```

```python
model.classes_
```

```python
model.predict_proba(X_test[:10])
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(GaussianNB(), X_train, y_train, cv=5)
```
