```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.data
```

```python
digits.target
```

Creating a new dataframe

```python
import pandas as pd
```

```python
df_digits = pd.DataFrame(digits.data,digits.target)
df_digits.head()
```

```python
df_digits['target'] =digits.target
df_digits.head(10)
```

Creating inputs(X), target(y) directly while splitting the data into train and test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_digits.drop('target',axis ='columns'), digits.target, test_size=0.2)
```

Training the model

```python
from sklearn.ensemble import RandomForestClassifier
model_digits = RandomForestClassifier()
```

```python
model_digits.fit(X_train, y_train)
```

```python
model_digits.score(X_test, y_test)
```

```python
model_digits.predict(X_test)
```

```python
model_digits = RandomForestClassifier(n_estimators=10)
```

```python
model_digits.fit(X_train, y_train)
```

```python
model_digits.score(X_test, y_test)
```

Generating confusion matrix to check where the model went wrong

```python
from sklearn.metrics import confusion_matrix
```

```python
y_predicted = model_digits.predict(X_test)
cm = confusion_matrix(y_test,y_predicted )
```

```python
import seaborn as sn
import matplotlib.pyplot as plt

plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
import pandas as pd
```

```python
df_iris = pd.DataFrame(iris.data, columns=iris.feature_names)
df_iris.head()
```

```python
df_iris['target'] = iris.target
df_iris.head()
```

Splitting data into test and train

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_iris.drop('target',axis ='columns'), iris.target, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model_iris = RandomForestClassifier()
```

```python
model_iris.fit(X_train, y_train)
```

```python
model_iris.score(X_test, y_test)
```

```python
y_predicted_iris = model_iris.predict(X_test)
```

```python
cm_iris = confusion_matrix(y_test,y_predicted_iris )
```

```python
plt.figure(figsize=(10,7))
sn.heatmap(cm_iris, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
model_iris = RandomForestClassifier(n_estimators=50)
```

```python
model_iris.fit(X_train, y_train)
```

```python
model_iris.score(X_test, y_test)
```
