```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('/content/drive/MyDrive/ml_practice/linear_regression/homeprices.csv')
df.head(3)
```

```python
## Plotting with data

plt.xlabel('area (sq ft)')
plt.ylabel('price ($)')
plt.scatter(df.area, df.price, color='red', marker='+')
```

```python
## assigning the model
reg = linear_model.LinearRegression()
```

```python
## Training the model
reg.fit(df[['area']], df.price) # learn the relationship, sklearn needs to differentiate between features and targets, features - 2D
```

```python
reg.predict([[5000]])
```

```python
df_2 = pd.read_csv('/content/drive/MyDrive/ml_practice/linear_regression/areas.csv')
df_2.head(3)
```

```python
p = reg.predict(df_2)
p
```

```python
df_2['prices'] = p
```

```python
plt.xlabel('area (sq ft)')
plt.ylabel('price ($)')
plt.scatter(df_2.area, df_2.prices, color='red', marker='+')
plt.plot(df_2.area, reg.predict(df_2[['area']]), color='blue')
```

```python
df_2.to_csv('/content/drive/MyDrive/ml_practice/linear_regression/prediction.csv', index = False)
```
