```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.data[0]
```

```python
plt.matshow(digits.images[0])
plt.gray()
```

```python
digits.target[0:5]
```

```python
digits.target[0]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_predicted = model.predict(X_test)
```

```python
model.score(X_test, y_test)
```

Confusion Matrix

Even though, the model is good because of its accuracy of 95%, we can still check where did it go wrong with Confusion Matrix - for large data it tells where it went wrong in a picture

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_predicted)
```

```python
cm
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True) # heatmap - turns numbers into colors to view patterns easily(large - brighter, smaller - lighter), annot - numbers inside the cell
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
iris.data[0]
```

```python
iris.target_names[0]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
iris_model = LogisticRegression()
```

```python
iris_model.fit(X_train, y_train)
```

```python
y_predicted = iris_model.predict(X_test)
```

```python
iris_model.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm_iris = confusion_matrix(y_test, y_predicted)
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm_iris, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Actual')
```
