Loading datasets from sklearn

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
iris.target
```

```python
iris.target_names
```

```python
import pandas as pd
import numpy as np
```

Creating a new dataframe with only needed columns

```python
df_iris = pd.DataFrame(iris.data, columns =iris.feature_names )
df_iris.head()
```

Adding target to new dataframe

```python
df_iris['target'] = iris.target
df_iris.head()
```

Adding human readable target name that maps to the its numerical value

```python
df_iris['flower_name'] = df_iris.target.apply(lambda x:iris.target_names[x])
df_iris.head()
```

```python
import matplotlib.pyplot as plt
```

Assigning each categorical target to new variable just to see scatter plots with respective to its features

```python
df0 = df_iris[df_iris.target == 0]
df1 = df_iris[df_iris.target == 1]
df2 = df_iris[df_iris.target == 2]
```

```python
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color = 'green', marker ='+')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color = 'blue', marker ='.')
```

```python
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'], color = 'green', marker ='+')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], color = 'blue', marker ='.')
```

```python
from sklearn.model_selection import train_test_split
```

Modifying inputs and target

```python
X = df_iris.drop(['target', 'flower_name'], axis = 'columns')
y = df_iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.svm import SVC
model_iris = SVC()
```

Training the model with Support Vector Classifier algorithm

```python
model_iris.fit(X_train, y_train)
```

```python
model_iris.predict(X_test)
```

```python
model_iris.score(X_test, y_test)
```

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

Loading datasets from sklearn

```python
dir(digits)
```

```python
digits.target
```

```python
digits.target_names
```

Creating a new dataframe with only needed columns

```python
df_digits = pd.DataFrame(digits.data,digits.target)
df_digits.head()
```

Adding target to new dataframe

```python
df_digits['target'] = digits.target
df_digits.head()
```

```python
df_digits['digit_name'] = df_digits.target.apply(lambda x:digits.target_names[x])
df_digits.head()
```

Modifying inputs and target

```python
X = df_digits.drop(['target','digit_name'], axis = 'columns')
X.head()
```

```python
y = df_digits.target
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.svm import SVC
model_digit = SVC()
```

Training the model with Support Vector Classifier algorithm

```python
model_digit.fit(X_train, y_train)
```

```python
model_digit.predict(X_test)
```

```python
model_digit.score(X_test, y_test)
```

Training and Testing the model with different parameters

```python
model_digit = SVC(kernel='linear')
```

```python
model_digit.fit(X_train, y_train)
```

```python
model_digit.score(X_test, y_test)
```

```python
model_digit = SVC(C=10)
```

```python
model_digit.fit(X_train, y_train)
```

```python
model_digit.score(X_test, y_test)
```
