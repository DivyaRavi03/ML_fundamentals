```python
df = pd.read_csv('/content/drive/MyDrive/ml_practice/feature_engineering/iqr/heights (2).csv')
df.head()
```

```python
df.describe()
```

```python
Q1 = df.height.quantile(0.25)
Q3= df.height.quantile(0.75)
Q1, Q3
```

```python
iqr = Q3 - Q1
iqr
```
