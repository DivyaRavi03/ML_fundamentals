Working with percentile rank with
## Complex Dataset

```python
df_prices = pd.read_csv('/content/drive/MyDrive/ml_practice/feature_engineering/percentile_rank/bhp.csv')
df_prices.head()
```

```python
df_prices.shape
```

```python
df_prices.describe()
```

```python
min_threshold, max_threshold = df_prices['price_per_sqft'].quantile([0.001, 0.999])
min_threshold, max_threshold
```

Below shown data can be legitimate sometimes but those are also outliers to our data causing issue in data analysis process thats why removing data is better.
Here we have used one of outlier detection method called as Percentile rank using quantile()

```python
df_prices[df_prices['price_per_sqft']<min_threshold]
```

```python
df_prices[df_prices['price_per_sqft']>max_threshold]
```

```python
df2 = df_prices[(df_prices['price_per_sqft']>min_threshold) & (df_prices['price_per_sqft']<max_threshold)]
df2.head()
```

Removed some of the outliers

```python
df2.shape
```

To print some random rows

```python
df2.sample(10)
```
