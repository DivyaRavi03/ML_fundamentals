```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/ml_practice/naive_bayes/spam.csv')
df.head()
```

```python
df.groupby('Category').describe()
```

Creatin a new column, where text column 'Category' is changed to new numerical column

```python
df['spam'] = df['Category'].apply(lambda x: 1 if x=='spam' else 0)
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.Message, df.spam)
```

Changing the sentence into words into numbers using CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
X_train_count = v.fit_transform(X_train.values)
X_train_count.toarray()[:5]
```

Using MultiNomial NB to train and predict

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count, y_train)
```

```python
prediction = v.transform(X_test)
model.predict(prediction)
```

```python
model.score(prediction, y_test)
```

Instead of using fit_transform every time to predict data, we can use Pipeline to do it in easier steps.

```python
from sklearn.pipeline import Pipeline
clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
clf.fit(X_train, y_train)
```

```python
clf.predict(X_test)
```

```python
clf.score(X_test, y_test)
```

```python
from sklearn.datasets import load_wine
wine = load_wine()
```

```python
dir(wine)
```

```python
df = pd.DataFrame(wine.data, columns = wine.feature_names)
df.head()
```

```python
df['target'] = wine.target
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(wine.data, wine.target, test_size =0.25)
```

```python
from sklearn.naive_bayes import GaussianNB
model_gaus = GaussianNB()
```

```python
model_gaus.fit(X_train, y_train)
```

```python
model_gaus.score(X_test, y_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
model_multi = MultinomialNB()
```

```python
model_multi.fit(X_train, y_train)
```

```python
model_multi.score(X_test, y_test)
```

Gaussian Naive Bayes performs better for this dataset compared to Multinomial Naive Bayes
