KFold Cross Validation - compares all machine learning algorithm after training the model with different data folds / batches and tells us which algorithm to use for the specific dataset with its high accuracy.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
import numpy as np
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.3)
```

```python
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)
lr_model.score(X_test, y_test)
```

```python
svc_model = SVC()
svc_model.fit(X_train, y_train)
svc_model.score(X_test, y_test)
```

```python
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
rf_model.score(X_test, y_test)
```

Instead of checking like above, we use Cross Validation for splitting the data equally as train and test into many folds to train and test model.

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=3)
kf
```

```python
def get_score(model, X_train, X_test, y_train, y_test):
  model.fit(X_train, y_train)
  return model.score(X_test, y_test)
```

We can also use Stratified KFold, because while shuffling it may take same set of data for training, which may affect testing, but SKF makes sure this mistake never happens

```python
scores_lr = []
scores_svc = []
scores_rf = []
```

```python
for train_index, test_index in kf.split(digits.data):
  X_train, X_test, y_train, y_test = digits.data[train_index], digits.data[test_index], \
                                     digits.target[train_index], digits.target[test_index]

  scores_lr.append(get_score(lr_model, X_train, X_test, y_train, y_test))
  scores_svc.append(get_score(svc_model, X_train, X_test, y_train, y_test))
  scores_rf.append(get_score(rf_model, X_train, X_test, y_train, y_test))
```

```python
scores_lr
```

```python
scores_svc
```

```python
scores_rf
```

Instead of using for loop to go through each time for same set of dataset, we can also use cross_val_score that does all of this calculations

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(LogisticRegression(), digits.data, digits.target)
```

```python
cross_val_score(SVC(), digits.data, digits.target)
```

```python
cross_val_score(RandomForestClassifier(), digits.data, digits.target)
```

From the above comparsion, Support Vector Classifier performs well with Digits data when compared with other algorithms such as Logistic Regression and Random Forest Classifier
